```python
import pandas as pd

df = pd.read_csv("data/mlbb_dataset.csv")

print(df.shape)
print(df.describe())
print(df["winner"].value_counts())
```

```python
import matplotlib
print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["gold_diff"], df["winner"])
plt.xlabel("Gold Difference")
plt.ylabel("Winner")
plt.show()

plt.scatter(df["kill_diff"], df["winner"])
plt.xlabel("Kill Difference")
plt.ylabel("Winner")
plt.show()

print(df.corr())

df["gold_per_min"] = df["gold_diff"] / df["minute"]
print(df.head())
```
