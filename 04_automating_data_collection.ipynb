```python
import cv2

video = cv2.VideoCapture("match1.mp4")

if video.isOpened():
    print("video loaded successfully")
else:
    print("error loading video")
```

checking if Python can jump through the video and capture frames automatically?

```python
fps = video.get(cv2.CAP_PROP_FPS)
print("Fps =",fps)

total_frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
print("Total Frames =", total_frames)

duration = total_frames / fps
print("Duration in seconds =", duration) 
```

jumping to specific frame

```python
video.set(cv2.CAP_PROP_POS_FRAMES, 2000)

success, frame = video.read()
# cv2.imwrite("frame2000.jpg", frame)
print("saved frame 2000")
```

calculating interval

```python
fps = video.get(cv2.CAP_PROP_FPS)
interval = int(fps * 60)
print(interval)
```

making folder from screenshot

```python
import os
os.makedirs('frames', exist_ok=True)
print("folder created")
```

loop through video

```python
frame_number = 0
count = 0

while True:
    #jump to frame
    video.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    #read frame
    success, frame = video.read()
    #stop if video ended
    if not success:
        break
    #save image
    filename = f"frames/frame_{count}.jpg"
    cv2.imwrite(filename, frame)
    print("saved", filename)
    #move to next interval
    frame_number += interval
    count += 1
```
