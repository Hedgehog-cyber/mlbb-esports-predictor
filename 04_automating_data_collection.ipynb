checking if video can be opened

```python
import cv2

video = cv2.VideoCapture("match1.mp4")

if video.isOpened():
    print("video loaded successfully")
else:
    print("error loading video")
```

```python
success, frame = video.read()

print(success)

cv2.imwrite("first_frame.jpg", frame)
print('saved')
```

checking if Python can jump through the video and capture frames automatically?

```python
fps = video.get(cv2.CAP_PROP_FPS)
print("Fps =",fps)

total_frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
print("Total Frames =", total_frames)

duration = total_frames / fps
print("Duration in seconds =", duration) 
```

jumping to specific frame

```python
video.set(cv2.CAP_PROP_POS_FRAMES, 2000)

success, frame = video.read()
cv2.imwrite("frame2000.jpg", frame)
print("saved frame 2000")
```
