```python
import pandas as pd

df = pd.read_csv("data/mlbb_dataset.csv")

df["gold_per_min"] = df["gold_diff"] / df["minute"]
```

```python
train_df = df[df["game_id"] <= 16]
test_df = df[df["game_id"] > 16]

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

features = [
    "minute",
    "kill_diff",
    "gold_diff",
    "gold_per_min",
    "turret_diff",
    "turtle_diff",
    "lord_diff"
]
```

```python
X_train = train_df[features]
y_train = train_df["winner"]

X_test = test_df[features]
y_test = test_df["winner"]

print(X_train.head())
print(y_train.head())
```

importing and training model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)

model.fit(X_train, y_train)

print("Model trained successfully")
```

```python
predictions = model.predict(X_test)

print(predictions)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, predictions)

print("Accuracy =", accuracy)
```

```python
probabilities = model.predict_proba(X_test)

print(probabilities)
```

```python
for feature, coef in zip(features, model.coef_[0]):
    print(feature, ":", round(coef, 4))
```

earlier mmodel not performing that good

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

pred = rf.predict(X_test)

from sklearn.metrics import accuracy_score

print("Accuracy:", accuracy_score(y_test, pred))

for feature, importance in zip(features, rf.feature_importances_):
    print(feature, ":", round(importance, 4))
```
